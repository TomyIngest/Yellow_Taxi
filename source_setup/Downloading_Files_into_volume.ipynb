```python
def download_to_volume(year):
    import urllib.request

    months = ['01','02','03','04','05','06','07','08','09','10','11','12']

    url_template = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_'
    directory_path = '/Volumes/ny_taxi/data/source_files/Taxi_data/'

    dbutils.fs.mkdirs(f"{directory_path}{str(year)}") 

    for month in months:
        working_url = f"{url_template}{str(year)}-{month}.parquet"

        # storing file from url into variable
        try:
            with urllib.request.urlopen(working_url) as url_open:
                file_read = url_open.read()
        except:
            break

        # storing file into directory
        with open(f"{directory_path}{str(year)}/NY_Taxi_{str(year)}-{month}.parquet", "wb") as file_opened:
            file_opened.write(file_read)

        

download_to_volume(2025)
```
