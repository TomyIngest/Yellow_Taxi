```python
# method which Use
    # 1. Urllib.request
    # 2. write 

import urllib.request

url = 'https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv'
directory_path = '/Volumes/ny_taxi/data/source_files/other/'

with urllib.request.urlopen(url) as url_opened:
    file_read = url_opened.read() # file_read read whole file into RAM

with open(f"{directory_path}taxi_zone_lookup.csv", "wb") as file_opened:
    file_opened.write(file_read)
```

```python
# modern method of writing files into directory from API
    # 1. using requests for connecting to API
    # 2. using shutil to copy from API


def download_to_volume(year):
    import shutil
    import requests

    months = ['01','02','03','04','05','06','07','08','09','10','11','12']

    url_template = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_'
    directory_path = '/Volumes/ny_taxi/data/source_files/Taxi_data/'

    dbutils.fs.mkdirs(f"{directory_path}{str(year)}") 

    for month in months:
        working_url = f"{url_template}{str(year)}-{month}.parquet"
        response =  requests.get(working_url, stream=True)

        if response.status_code == 403: # If URL can not be connected then break the loop
            break
        with (
            response,
            open(f"{directory_path}{str(year)}/NY_Taxi_{str(year)}-{month}.parquet", "wb") as file_opened
        ):
            shutil.copyfileobj(response.raw,file_opened)
  
download_to_volume(2026)
```
